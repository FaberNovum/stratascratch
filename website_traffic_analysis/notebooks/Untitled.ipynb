## Question 5

**How does the clickrate distribute across different links?**

```python
# need dataframe grouped by linkid with clickrate aggregation
# drop preview events
# then get relative frequency of remaining even
```

```python
df['event'].unique()
```

```python
new_df = df[['event', 'linkid']]
```

```python
# drop previews
new_df = new_df[new_df['event'] != 'preview']
```

```python
new_df
```

```python
grouped_clicks_df = pd.DataFrame(new_df.groupby('event').value_counts()['click'])
```

```python
grouped_clicks_df.reset_index(inplace=True)
```

```python
grouped_clicks_df.rename(columns={0: "clicks"}, inplace=True)
```

```python
grouped_clicks_df
```

```python
grouped_pageviews_df = pd.DataFrame(new_df.groupby('event').value_counts()['pageview'])
```

```python
grouped_pageviews_df.reset_index(inplace=True)
```

```python
grouped_pageviews_df.rename(columns={0: "pageviews"}, inplace=True)
```

```python
grouped_pageviews_df
```

```python
merged_df = grouped_clicks_df.merge(grouped_pageviews_df, on='linkid')
```

```python
merged_df
```

```python
merged_df['clickrate'] = merged_df['clicks'] / merged_df['pageviews']
```

```python
grouped_clickrate_df = pd.DataFrame(merged_df.groupby('linkid')['clickrate'].sum())
```

```python
grouped_clickrate_df.reset_index(inplace=True)
```

```python
grouped_clickrate_df
```

```python
# confirm with first entry
grouped_clickrate_df[grouped_clickrate_df['linkid'] == '2d896d31-97b6-4869-967b-1c5fb9cd4bb8']
```

```python
import matplotlib.pyplot as plt
```

```python
# now plot
ax = grouped_clickrate_df["clickrate"].plot(kind="hist", figsize=(7, 7), bins=100)
_ = ax.set_xlabel("Click Rate")
```

```python
# because the data is highly skewed, it makes more sense to plot the log of it
grouped_clickrate_df["click_rate_log2"] = np.log2(grouped_clickrate_df["clickrate"])
ax = grouped_clickrate_df["click_rate_log2"].plot(kind="hist", figsize=(7, 7), bins=100)
_ = ax.set_xlabel("Log2(Click Rate)")
```

```python
ax = grouped_clickrate_df.boxplot(column="click_rate_log2")
_ = ax.set_ylabel("Value")
```
